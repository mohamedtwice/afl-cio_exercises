```python
# Importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
# Convert spreadsheets into variables and merge them
v12 = pd.read_excel('../data/Volunteers 1.1.12-12.31.12.xlsx')
v14 = pd.read_excel('../data/Volunteers 1.1.14-12.31.14.xlsx')
v12_14 = pd.concat([v12, v14])
```

```python
# Create a new dataframe that has only Completed shifts
completed = v12_14[v12_14['Status'] == 'Completed']
completed
```

```python
# Create a new dataframe that has only includes our selected roles (Walks, Phone Banks, Data Entry/Support, GOTV)
completed_voltypes = completed[(completed['Role'] == 'Walks') | (completed['Role'] == 'Phone Banks') | (completed['Role'] == 'Data Entry/Support') | (completed['Role'] == 'GOTV')]
completed_voltypes
```
