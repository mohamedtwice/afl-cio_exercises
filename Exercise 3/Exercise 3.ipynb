```python
# Importing pandas, numpy, and xlsxwriter
import pandas as pd
import numpy as np
from xlsxwriter.utility import xl_rowcol_to_cell
```

```python
# Convert spreadsheets into variables and merge them
v12 = pd.read_excel('../data/Volunteers 1.1.12-12.31.12.xlsx')
v14 = pd.read_excel('../data/Volunteers 1.1.14-12.31.14.xlsx')
v12_14 = pd.concat([v12, v14])
```

```python
# Replace spaces with underscores to make it easier to work with column names
v12.columns = v12.columns.str.replace(' ', '_')
v14.columns = v14.columns.str.replace(' ', '_')
v12_14.columns = v14.columns.str.replace(' ', '_')
```

```python
# I want to only look at volunteer shifts that have been completed 
completed12 = v12[v12['Status'] == 'Completed']
completed14 = v14[v14['Status'] == 'Completed']
completed12_14 = v12_14[v12_14['Status'] == 'Completed']
```

```python
# Here I'm looking at the MyCampaignID column and looking at how many unique values exist so I can understand how many people volunteered.
completed12.MyCampaignID.unique().shape, completed14.MyCampaignID.unique().shape, completed12_14.MyCampaignID.unique().shape
```

```python
# Looks at each cycle and finds the number of volunteers each state has
completed12.groupby(['Home_State']).size(), completed14.groupby(['Home_State']).size()
```

```python
# Here I'm looking each cycle and finds the number of volunteers each affiliate had
completed12.groupby(['International_Name']).size(), completed14.groupby(['International_Name']).size()
```

```python
# Here I'm looking each state and displaying the number of volunteers each affiliate had in 2012
completed12.groupby(['Home_State', 'International_Name']).size()
```

```python
# Here I'm looking each state and displaying the number of volunteers each affiliate had in 2014
completed14.groupby(['Home_State', 'International_Name']).size()
```

```python
# Here I'm looking each affiliate and displaying osme information about each role they filled(the count) in 2012
completed12.groupby(['International_Name', 'Role']).describe()
```

```python
# Here I'm looking each affiliate and displaying osme information about each role they filled(the count) in 2014
completed14.groupby(['International_Name', 'Role']).describe()
```

```python
# Create an excel file from the final dataframes, results.
completed12_writer = pd.ExcelWriter('Exercise3-12.xlsx', engine='xlsxwriter')
completed14_writer = pd.ExcelWriter('Exercise3-14.xlsx', engine='xlsxwriter')
completed12.to_excel(completed12_writer, index=False, sheet_name='report')
completed14.to_excel(completed14_writer, index=False, sheet_name='report')
completed12_writer.save()
completed14_writer.save()
```
