```python
import pandas as pd
import numpy as np
import xlsxwriter
```

```python
v12 = pd.read_csv('Volunteers 1.1.12-12.31.12.csv')
v14 = pd.read_csv('Volunteers 1.1.14-12.31.14.csv')
v12_14 = pd.concat([v12, v14])
```

```python
v12_14.columns
```

```python
v12_14.columns = v12_14.columns.str.replace(' ', '_')
```

```python
v12_14.columns
```

```python
v12_14['Home_State'].replace(['Hollywood'],['FL'],inplace=True)
v12_14['Home_State'].replace(['Ft Lauderdale'], ['FL'],inplace=True)
v12_14['Home_State'].replace(['Sunrise'], ['FL'],inplace=True)
```

```python
v12_14[('Home_State')].unique()
```

```python
fl_file = v12_14[v12_14['Home_State'].isin(['FL'])].copy()
nc_file = v12_14[v12_14['Home_State'].isin(['NC'])].copy()
oh_file = v12_14[v12_14['Home_State'].isin(['OH'])].copy()
wi_file = v12_14[v12_14['Home_State'].isin(['WI'])].copy()
mo_file = v12_14[v12_14['Home_State'].isin(['MO'])].copy()
pa_file = v12_14[v12_14['Home_State'].isin(['PA'])].copy()
nv_file = v12_14[v12_14['Home_State'].isin(['NV'])].copy()
```

```python
nv_file.tail()
```

```python
fl_writer = pd.ExcelWriter('fl_file.xlsx', engine='xlsxwriter')
nc_writer = pd.ExcelWriter('nc_file.xlsx', engine='xlsxwriter')
oh_writer = pd.ExcelWriter('oh_file.xlsx', engine='xlsxwriter')
wi_writer = pd.ExcelWriter('wi_file.xlsx', engine='xlsxwriter')
mo_writer = pd.ExcelWriter('mo_file.xlsx', engine='xlsxwriter')
pa_writer = pd.ExcelWriter('pa_file.xlsx', engine='xlsxwriter')
nv_writer = pd.ExcelWriter('fl_file.xlsx', engine='xlsxwriter')
```

```python
fl_file.to_excel(fl_writer, sheet_name='Sheet1')
nc_file.to_excel(nc_writer, sheet_name='Sheet1')
oh_file.to_excel(oh_writer, sheet_name='Sheet1')
wi_file.to_excel(wi_writer, sheet_name='Sheet1')
mo_file.to_excel(mo_writer, sheet_name='Sheet1')
pa_file.to_excel(pa_writer, sheet_name='Sheet1')
nv_file.to_excel(nv_writer, sheet_name='Sheet1')
```

```python
fl_writer.save()
nc_writer.save()
oh_writer.save()
wi_writer.save()
mo_writer.save()
pa_writer.save()
nv_writer.save()
```
